```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Cricket_Matches_Data.csv")
df.head()
```

```python
for i in df:
    print(i,":",df[i].unique())
```

# Matches per Team

```python
Teams = pd.concat([df['team1'], df['team2']])
matches_per_team0 = Teams.value_counts()
matches_per_team = pd.DataFrame(matches_per_team0, columns=["matches_per_team"])
matches_per_team = matches_per_team.reset_index()
mpt = matches_per_team.rename(columns={'index': 'team'})
mpt
```

```python
plt.barh(mpt['team'], mpt['matches_per_team'])
plt.xlabel("No of Matches")
plt.ylabel("Teams")
plt.title("No of Matches per Team")
plt.show()
```

# Matches per Ground

```python
no_of_matches_per_ground0 = df["venue"].value_counts()
no_of_matches_per_ground = pd.DataFrame(no_of_matches_per_ground0)
no_of_matches_per_ground = no_of_matches_per_ground.reset_index()
mpg = no_of_matches_per_ground.rename(columns={'index': 'ground', 'venue':'no_of_matches'})
mpg
```

```python
plt.barh(mpg['ground'], mpg['no_of_matches'])
plt.xlabel("No of Matches")
plt.ylabel("Grounds")
plt.title("No of Matches per Ground")
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```

# Number of Grounds

```python
no_of_ground = []
no_of_ground.append(no_of_matches_per_ground.shape[0])
no_of_ground
```

# Matches per Year

```python
no_of_matches_per_year0 = df["season"].value_counts()
no_of_matches_per_year = pd.DataFrame(no_of_matches_per_year0)
no_of_matches_per_year = no_of_matches_per_year.reset_index()
mpy = no_of_matches_per_year.rename(columns={'index': 'year', 'season':'no_of_matches'})
mpy
```

```python
plt.barh(mpy['year'], mpy['no_of_matches'])
plt.xlabel("No of Matches")
plt.ylabel("Years")
plt.title("No of Matches per Year")
plt.show()
```

# Wins per Team

```python
no_of_wins_per_team0 = df["winner"].value_counts()
no_of_wins_per_team = pd.DataFrame(no_of_wins_per_team0)
no_of_wins_per_team = no_of_wins_per_team.reset_index()
wpt = no_of_wins_per_team.rename(columns={'index': 'team', 'winner':'no_of_wins'})
wpt
```

```python
plt.barh(wpt['team'], wpt['no_of_wins'])
plt.xlabel("No of Wins")
plt.ylabel("Teams")
plt.title("No of Wins per Team")
plt.show()
```

# Venue vs Toss Decision

```python
ground_vs_toss_decision = df.groupby(['venue', 'toss_decision']).size()
ground_dict = {}
for ground, toss_decision in ground_vs_toss_decision.index:
    count = ground_vs_toss_decision[(ground, toss_decision)]
    if ground not in ground_dict:
        ground_dict[ground] = {}
    ground_dict[ground][toss_decision] = count
print(ground_dict)
```

# Players of Matchs per Player

```python
Number_of_Players_of_Match_per_Player0 = df["player_of_match"].value_counts()
Number_of_Players_of_Match_per_Player = pd.DataFrame(Number_of_Players_of_Match_per_Player0)
Number_of_Players_of_Match_per_Player = Number_of_Players_of_Match_per_Player.reset_index()
pmp = Number_of_Players_of_Match_per_Player.rename(columns={'index': 'player'})
pmp
```

```python
l = {}
for j in range(pmp.shape[0]-1):
    l[pmp["player"][j]] = pmp["player_of_match"][j]
print(l)        
```

# Matches per Empires

```python
Empires = pd.concat([df['umpire1'], df['umpire2'], df['umpire3']])
```

```python
Empires = Empires[~Empires.isnull()]
```

```python
Number_of_empires_per_match0 = Empires.value_counts()
Number_of_empires_per_match = pd.DataFrame(Number_of_empires_per_match0)
Number_of_empires_per_match = Number_of_empires_per_match.reset_index()
epm = Number_of_empires_per_match.rename(columns={'index':'empires', 0:'no_of_matches'})
epm
```

```python
plt.barh(epm['empires'], epm['no_of_matches'])
plt.xlabel("Empires")
plt.ylabel("No of Matches")
plt.title("No of Matches per Empire")
fig = plt.gcf()
fig.set_size_inches(16,16)
plt.show()
```

# Tosses Won per Team

```python
Tosses_won_per_team0 = df["toss_winner"].value_counts()
Tosses_won_per_team = pd.DataFrame(Tosses_won_per_team0)
Tosses_won_per_team = Tosses_won_per_team.reset_index()
twt = Tosses_won_per_team.rename(columns={'index': 'team', 'toss_winner':'no_of_tosses_won'})
twt
```

```python
plt.barh(twt['team'], twt['no_of_tosses_won'])
plt.xlabel("No of Tosses Won")
plt.ylabel("Team")
plt.title("No of Tosses Won per Team")
plt.show()
```

# Team vs Toss Decision

```python
team_vs_toss_decision = df.groupby(['toss_winner', 'toss_decision']).size()
team_vs_toss_decision = team_vs_toss_decision.reset_index(name='count').pivot(index='toss_winner', columns='toss_decision', values='count')
team_vs_toss_decision.plot(kind='bar', width = 0.8)
plt.xlabel('Team')
plt.ylabel('Count')
plt.title('Team vs Toss Decision')
plt.show()
```
